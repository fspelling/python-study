```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_gas = pd.read_csv('data.csv')
df_gas
```

```python
plt.plot(df_gas['Year'], df_gas['Australia'], 'b.-', label='Australia')
plt.plot(df_gas['Year'], df_gas['Italy'], 'g.-', label='Italy')
plt.plot(df_gas['Year'], df_gas['USA'], 'r.-', label='USA')

plt.xticks(df_gas['Year'][::2])

plt.title('Vendas anuais de gas em $(USA)')
plt.xlabel('Anos')
plt.ylabel('Preco')

plt.legend()
plt.show()
```

```python
paises_list = ['Australia', 'Italy', 'USA']

for column_gas, series_gas in df_gas.items():
    if column_gas in paises_list:
        plt.plot(df_gas['Year'], series_gas, label=column_gas, marker='.')

plt.title('Vendas anuais de gas em $(USA)')
plt.xlabel('Anos')
plt.ylabel('Preco')

plt.xticks(df_gas['Year'][::2])

plt.legend()
# plt.show()

plt.savefig('result_grafico.pdf')
```
