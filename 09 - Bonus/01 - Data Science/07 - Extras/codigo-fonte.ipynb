```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mouths = [0, 1, 2]
clients = { 'Arthur': [10, 30, 20], 'Jose': [20, 10, 50], 'Fabio': [30, 40, 80], 'Larissa': [40, 90, 10] }
data_clients = { 'Mouths':  mouths }

for name_client, values_client in clients.items():
    data_clients[name_client] = values_client
```

```python
df_clients = pd.DataFrame(data=data_clients)
df_clients
```

```python
plot_column_list = [col_series for col_name, col_series in df_clients.items() if col_name != 'Mouths']

for col_series_plot in plot_column_list:
    plt.plot(mouths, col_series_plot, marker='.', label=col_series_plot.name)

plt.title('Ranking trimestral de clientes com mais compras')
plt.xlabel('Meses')
plt.ylabel('Compras($)')
plt.xticks(mouths, ['Jan', 'Fev', 'Mar'])
plt.legend()

plt.show()
```
